###### Create a function to sum two numbers

```python
def sum_of_two_num(a,b):
    c = a+b
    return c
```

```python
sum_of_two_num(3,4)
```

```python
sum_of_two_num(6,9)
```

```python
def marks_percentage(E,H,M,S,SST):
    avg = (E+H+M+S+SST)/5
    print('Average Marks : ' , avg)
```

```python
marks_percentage(18,19,18,17,18)
```

###### Default Parameter Values

After creating a Function when there is no Parameter value , in that condition we can give a Defualt Paramter Value while creating a function.


```python
def designation_salary(name,designation, salary):
    print(name , 'working as ' , designation , ' and his salary is ', salary)
```

```python
designation_salary('Vijay', 'Business Analyst')
```

```python
def designation_salary(name= 'Not Given',designation = 'Analyst', salary= 'Not Given'):
        print(name , 'working as ' , designation , ' and his salary is ', salary)
```

```python
designation_salary('Sunil')
```
